# Encoding

```python
df["Agricultural"] = np.where(df["Land Cover"] == "Agricultural" , 1,0)
df["Desert"] = np.where(df["Land Cover"] == "Desert" , 1,0)
df["Forest"] = np.where(df["Land Cover"] == "Forest" , 1,0)
df["Urban"] = np.where(df["Land Cover"] == "Urban" , 1,0)
df["Water"] = np.where(df["Land Cover"] == "Water Body" , 1,0)

df.drop(columns=['Land Cover'],axis = 1, inplace = True)
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df["Peat"] = np.where(df["Soil Type"] == "Peat" , 1,0)
df["Loam"] = np.where(df["Soil Type"] == "Loam" , 1,0)
df["Sandy"] = np.where(df["Soil Type"] == "Sandy" , 1,0)
df["Clay"] = np.where(df["Soil Type"] == "Clay" , 1,0)
df["Silt"] = np.where(df["Soil Type"] == "Silt" , 1,0)

df.drop(columns=['Soil Type'],axis = 1, inplace = True)
df.head()
```

# Spliting

```python
x = df.drop(['Flood Occurred'], axis = 1)
y = df['Flood Occurred']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

# Scaling

```python
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

# NEURAL NETWORK

```python
# Define the MLP model with appropriate hyperparameters
mlp_model = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=1000, random_state=0)

# Train the model
mlp_model.fit(x_train, y_train)
```

```python
# Make predictions
train_mlp_pred = mlp_model.predict(x_train)
test_mlp_pred = mlp_model.predict(x_test)

# Accuracy Score
train_accuracy_mlp = accuracy_score(y_train, train_mlp_pred)
test_accuracy_mlp = accuracy_score(y_test, test_mlp_pred)

# Precision Score
train_precision_mlp = precision_score(y_train, train_mlp_pred)
test_precision_mlp = precision_score(y_test, test_mlp_pred)

# F1 Score
train_f1_mlp = f1_score(y_train, train_mlp_pred)
test_f1_mlp = f1_score(y_test, test_mlp_pred)

# Recall Score
train_recall_mlp = recall_score(y_train, train_mlp_pred)
test_recall_mlp = recall_score(y_test, test_mlp_pred)

# Confusion Matrix
train_mlp_cm = confusion_matrix(y_train, train_mlp_pred)
test_mlp_cm = confusion_matrix(y_test, test_mlp_pred)

# Print results
print(f"Train Accuracy: {train_accuracy_mlp}")
print(f"Test Accuracy: {test_accuracy_mlp}")
print(f"Train Precision: {train_precision_mlp}")
print(f"Test Precision: {test_precision_mlp}")
print(f"Train F1 Score: {train_f1_mlp}")
print(f"Test F1 Score: {test_f1_mlp}")
print(f"Train Recall: {train_recall_mlp}")
print(f"Test Recall: {test_recall_mlp}")

print("\nConfusion Matrix (Train):")
print(train_mlp_cm)

print("\nConfusion Matrix (Test):")
print(test_mlp_cm)
```
